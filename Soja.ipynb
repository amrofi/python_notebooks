# Regressão

Usaremos os pacotes numpy, pandas, sklearn.linear_model, statsmodels.formula.api.

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.formula.api as sm
eq1 = sm.ols(formula="QSOJA~FERTILIZANTE+TRATOR+MO", data=dados).fit()
print(eq1.summary())
```

Portanto, nosso modelo tem um $R^2 = 0.465$, razoavelmente baixo, indicando que apenas 46% das variações de QSOJA são explicadas pelas variáveis explicativas.  

Os coeficientes do intercepto, de TRATOR e de MO foram estatisticamente significativos, respectivamente a 1%, 1% e 10% de significância. Ou seja, não foi possível dizer que FERTILIZANTE afeta QSOJA.
